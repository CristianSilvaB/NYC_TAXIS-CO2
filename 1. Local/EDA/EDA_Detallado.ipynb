```python
def data_characterics(dataframe):
    list =[]
    for columna in dataframe.columns:
      type_data = dataframe[columna].dtypes
      data_null = dataframe[columna].isnull().sum()
      porcentaje = round((data_null / len(dataframe)) * 100)
    
      list.append({'Columna':columna,
                   'Tipo':type_data,
                   'Nulos':data_null,
                    'Porcentaje':f'{porcentaje} %'} )
      
    return pd.DataFrame(list)
  
#Funcion para agregar etiqueta de magnitud a graficos de barras
def ubication_label(grafico):
  for p in grafico.patches:    
    height = p.get_height() # Obtiene la altura (frecuencia) de la barra
    pos_inicial_bar = p.get_x()
    ancho_barra = p.get_width() / 1.8
    
    x_pos = pos_inicial_bar + ancho_barra #Posicion donde se coloca la etiqueta
    y_pos = height # Ajuste de la posición vertical
    
    height = int(height)
        # Agrega la etiqueta 
    grafico.annotate(f'{height}', (x_pos, y_pos), ha='center', va='bottom')

data_characterics(taxis2023)
```

```python
filtrado = taxis2023[taxis2023['passenger_count'] != 0.0]
pasajeros =filtrado['passenger_count'].value_counts()

plt.figure(figsize=(8,5))
graphic = sns.barplot(x=pasajeros.index, y=pasajeros.values,
                      hue=pasajeros.index, palette='viridis')
ubication_label(graphic)
plt.title('Frecuencia de pasajeros por viaje')
plt.xlabel('Cantidad de pasajeros')
plt.show()
```

Se observa que los viajes tienen entre 1 y 9 pasajeros. La gran mayoría tiene solo 1 pasajero, con una frecuencia ampliamente superior al resto, totalizando 1,431,977 viajes. Le siguen los viajes con 2 pasajeros, con un total de 227,189, una frecuencia mucho menor. En el último puesto se encuentran los viajes con 9 pasajeros, con tan solo 15.

```python
servicio = taxis2023['Servicio'].value_counts()
colores = ['#E9C46A','#8CB369']
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
barras = sns.barplot(x=servicio.index, y=servicio.values, palette=colores, width=0.5)
ubication_label(barras)
plt.suptitle('Frecuencia de tipo de servicio')

plt.subplot(1,2,2)
plt.pie(servicio, autopct='%1.1f%%', startangle=90, labels=['Yellow','Green'], colors=colores)
plt.show()
```

El tipo de servicio mas demandados para el año 2023 es el de taxis Yellow con mas de 1 millon de viajes realizados. Este tipo de servicio representa el 59,9% del dataset. El servicio de taxis Green quedo en segundo lugar con mas de 700 mil viajes realizados.

```python
zona = taxis2023['Pickup_borough'].value_counts()
plt.figure(figsize=(8,5))
graphic = sns.barplot(x=zona.index, y=zona.values, width=0.5, palette='viridis')
ubication_label(graphic)
plt.title('Frecuencia de viajes por Distrito')
plt.show()
```

```python
#Se toma una muestra aleatoria de 10000 datos
muestra = taxis2023.sample(frac=0.01, random_state=42)
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
sns.scatterplot(y='trip_distance', x='total_amount', data=muestra)
plt.subplot(1,2,2)
#Con 'hue' diferenciamos el tipo de servicio
sns.scatterplot(y='trip_distance', x='total_amount', data=muestra, hue='Servicio', palette=colores)
plt.suptitle('Relacion entre trip_distance y total_amount ')
plt.show()
```

Al observar el grafico de dispersion, notamos que hay una correlacion positiva, a medida que la distancia recorrida aumenta, el precio del viaje tambien. En conclusion hay una relacion fuerte entre "trip_distance" y "total_amount".

Para el segundo gráfico en el cual los puntos están divididos según el tipo de servicio, la distribucion es similar, no se observa una diferencia entre el precio y el tipo de servicio.
