Eliminamos columnas inecesarias para nuestros objetivos

```python
# Lista de columnas que quieres eliminar
drops = ['barrels08', 'barrelsA08', 'charge240', 'city08U', 'cityA08', 'cityA08U','cityCD', 'cityE', 'cityUF', 'co2A',
         'co2TailpipeAGpm', 'co2TailpipeGpm', 'comb08', 'comb08U','combA08', 'combA08U', 'combE', 'combinedCD', 'combinedUF',
         'cylinders', 'displ', 'drive', 'engId', 'eng_dscr', 'feScore', 'fuelCost08', 'fuelCostA08', 'ghgScore', 'ghgScoreA',
         'highway08', 'highway08U', 'highwayA08', 'highwayA08U', 'highwayCD', 'highwayE', 'highwayUF','hlv', 'hpv', 'id', 'lv2',
         'lv4', 'mpgData', 'phevBlended', 'pv2', 'pv4', 'range', 'rangeCity', 'rangeCityA', 'rangeHwy','rangeHwyA', 'trany',
         'UCity', 'UCityA', 'UHighway', 'UHighwayA', 'youSaveSpend', 'guzzler', 'trans_dscr', 'tCharger','sCharger', 'atvType',
         'rangeA', 'evMotor', 'mfrCode', 'c240Dscr', 'charge240b', 'c240bDscr', 'createdOn', 'modifiedOn','startStop',
         'phevCity', 'phevHwy', 'phevComb', 'fuelType2', 'fuelType']

# Eliminar las columnas
VFED_nuevo = VFED.drop(columns=drops)
```

Entre los distintos tipos de combustible decidi dejar "fuelType1" porque nos da mayor informacion y las otras columnas tenian mas valores nulos

```python
VFED_nuevo.head(2)
```

```python
# Calcular el porcentaje de valores nulos en cada columna
porcentaje_nulos = (VFED_nuevo.isnull().mean()) * 100

# Imprimir el porcentaje de valores nulos en cada columna
print("Porcentaje de valores nulos en cada columna:")
print(porcentaje_nulos)
```

```python
# Encontrar valores nulos
valores_nulos = VFED_nuevo.isnull()

# Mostrar los valores nulos
print("Valores nulos en el archivo:")
print(valores_nulos)
```

```python
vfed_limpio = VFED_nuevo.dropna()
print("¿Hay filas vacías en el DataFrame?:", vfed_limpio.empty)
```

```python
vfed_limpio
```

```python
# Calcular el porcentaje de valores nulos en cada columna
porcentaje_nulos = (vfed_limpio.isnull().mean()) * 100

# Imprimir el porcentaje de valores nulos en cada columna
print("Porcentaje de valores nulos en cada columna:")
print(porcentaje_nulos)
```

```python
# Verificar si existen valores duplicados
existen_duplicados = vfed_limpio.duplicated().any()
# Imprimir el resultado
if existen_duplicados:
    print("Existen valores duplicados en el DataFrame.")
else:
    print("No existen valores duplicados en el DataFrame.")
```

```python
# Encontrar filas duplicadas basadas en todas las columnas
duplicados = vfed_limpio[vfed_limpio.duplicated(keep=False)]
duplicados
```

```python
# Eliminar todas las filas duplicadas basadas en todas las columnas
vfed_limpios = vfed_limpio.drop_duplicates()
vfed_limpios
```

Vamos a cambiar el nombre de nuestras columnas

```python
nuevos_nombres = {
    'fuelType1': 'Tipo de Combustible',
    'VClass': 'Categoria',
    'co2': 'CO2',
    'city08': 'City08'
}
vfed_final= vfed_limpios.rename(columns=nuevos_nombres)
vfed_final
```

Por ultimo vemos si los tipos de datos nuevamente

```python
tipos_de_datos = vfed_final.dtypes
print(tipos_de_datos)
```

Guardamos en formato parquet

```python
vfed_final.to_parquet('D:/escritorio/NYC-TAXIS/NYC_TAXIS-CO2/Data/Normalized/vfed_limpio.parquet', index=False)
```
